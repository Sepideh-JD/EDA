```python
import numpy as np
from collections import defaultdict
import os
import cv2

# Initialize a counter dictionary
class_counts = defaultdict(int)

# Go through all masks in mini_maps
for fname in os.listdir("mini_maps"):
    if fname.endswith(".png"):
        path = os.path.join("mini_maps", fname)
        mask = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

        unique, counts = np.unique(mask, return_counts=True)

        for u, c in zip(unique, counts):
            class_counts[u] += c

# Print results
print("Pixel counts per class across all masks:")
for k in sorted(class_counts):
    print(f"Class {k}: {class_counts[k]} pixels")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
from collections import defaultdict

# ---- Step 1: Paths (adjusted for mini dataset) ----
train_dir = "/content/mini_train"
map_dir = "/content/mini_maps"

# ---- Step 2: List images ----
train_images = sorted(os.listdir(train_dir))
map_images = sorted(os.listdir(map_dir))

print(f"Total train images: {len(train_images)}")
print(f"Total map images: {len(map_images)}")

# ---- Step 3: Show random image-mask pairs ----
def show_image_pair(idx):
    img = Image.open(os.path.join(train_dir, train_images[idx]))
    mask = Image.open(os.path.join(map_dir, map_images[idx]))

    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img)
    ax[0].set_title("Train Image")
    ax[0].axis('off')
    ax[1].imshow(mask, cmap='gray')
    ax[1].set_title("Mask")
    ax[1].axis('off')
    plt.tight_layout()
    plt.show()

# Show 3 random image pairs
for i in np.random.choice(len(train_images), 3, replace=False):
    show_image_pair(i)

# ---- Step 4: Check basic stats for images ----
def image_stats(image_path):
    img = np.array(Image.open(image_path))
    return {
        "shape": img.shape,
        "min": np.min(img),
        "max": np.max(img),
        "mean": np.mean(img),
        "std": np.std(img)
    }

print("\n📊 Image stats:")
for i in range(min(3, len(train_images))):
    stats = image_stats(os.path.join(train_dir, train_images[i]))
    print(f"{train_images[i]} --> {stats}")

# ---- Step 5: Count black masks ----
def check_black_image(path):
    img = np.array(Image.open(path))
    return np.all(img == 0)

black_count = sum(check_black_image(os.path.join(map_dir, f)) for f in map_images)
print(f"\n🕳️ Completely black masks: {black_count} / {len(map_images)}")

# ---- Step 6: Class distribution in masks ----
class_counts = defaultdict(int)

for fname in map_images:
    mask = np.array(Image.open(os.path.join(map_dir, fname)))
    unique, counts = np.unique(mask, return_counts=True)
    for val, count in zip(unique, counts):
        class_counts[val] += count

# Print counts
print("\n🧬 Pixel counts per class across masks:")
for k in sorted(class_counts):
    print(f"Class {k}: {class_counts[k]} pixels")

# Plot bar chart
plt.figure(figsize=(8, 4))
plt.bar(class_counts.keys(), class_counts.values(), color='slateblue')
plt.xlabel("Gleason Class Label")
plt.ylabel("Pixel Count")
plt.title("Distribution of Gleason Grades in Mask Annotations")
plt.xticks(list(class_counts.keys()))
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
