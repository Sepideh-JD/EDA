```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

print("=== Gleason 2019 Dataset Initial Exploration ===\n")

# Step 1: Check directory structure and available files
def explore_directory_structure():
    """Explore the dataset directory structure"""
    print("1. DIRECTORY STRUCTURE ANALYSIS")
    print("-" * 40)

    # List all directories and files in current path
    current_files = os.listdir('.')
    print(f"Files/folders in current directory: {len(current_files)}")
    for item in sorted(current_files):
        if os.path.isdir(item):
            print(f"📁 {item}/")
            # List contents of each directory
            try:
                sub_items = os.listdir(item)
                print(f"   Contains {len(sub_items)} items")
                if len(sub_items) <= 10:  # Show contents if not too many
                    for sub_item in sorted(sub_items)[:5]:
                        print(f"   - {sub_item}")
                    if len(sub_items) > 5:
                        print(f"   ... and {len(sub_items)-5} more items")
            except PermissionError:
                print("   (Permission denied)")
        else:
            print(f"📄 {item}")
    print()

# Step 2: Identify image and mask files
def identify_dataset_files():
    """Identify and match image files with their corresponding masks"""
    print("2. DATASET FILES IDENTIFICATION")
    print("-" * 40)

    # Common directory names for Gleason dataset
    possible_dirs = ['Train_imgs', 'Train', 'Images', 'Maps4_T', 'Mask', 'masks', 'labels']

    image_dirs = []
    mask_dirs = []

    for dir_name in os.listdir('.'):
        if os.path.isdir(dir_name):
            if any(keyword in dir_name.lower() for keyword in ['img', 'image', 'train']):
                image_dirs.append(dir_name)
            elif any(keyword in dir_name.lower() for keyword in ['map', 'mask', 'label']):
                mask_dirs.append(dir_name)

    print(f"Potential image directories: {image_dirs}")
    print(f"Potential mask directories: {mask_dirs}")

    # Get file lists from each directory
    all_files = {}
    for directory in image_dirs + mask_dirs:
        try:
            files = os.listdir(directory)
            all_files[directory] = files
            print(f"\n{directory}/ contains {len(files)} files")

            # Show file extensions
            extensions = {}
            for file in files:
                ext = os.path.splitext(file)[1].lower()
                extensions[ext] = extensions.get(ext, 0) + 1
            print(f"File extensions: {extensions}")

            # Show sample filenames
            print("Sample files:")
            for file in sorted(files)[:3]:
                print(f"  - {file}")

        except Exception as e:
            print(f"Error reading {directory}: {e}")

    return image_dirs, mask_dirs, all_files

# Step 3: Match images with their corresponding masks
def match_images_and_masks(image_dirs, mask_dirs, all_files):
    """Match image files with their corresponding mask files"""
    print("\n3. MATCHING IMAGES AND MASKS")
    print("-" * 40)

    if not image_dirs or not mask_dirs:
        print("❌ Could not find both image and mask directories")
        return []

    # Extract base IDs from filenames
    def extract_id(filename):
        """Extract the base ID from filename (remove extensions and suffixes)"""
        base = os.path.splitext(filename)[0]
        # Remove common suffixes
        for suffix in ['_classimg_nonconvex', '_mask', '_label', '_img']:
            if base.endswith(suffix):
                base = base[:-len(suffix)]
        return base

    # Get IDs from images and masks
    image_ids = set()
    mask_ids = set()

    for img_dir in image_dirs:
        if img_dir in all_files:
            for file in all_files[img_dir]:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                    image_ids.add(extract_id(file))

    for mask_dir in mask_dirs:
        if mask_dir in all_files:
            for file in all_files[mask_dir]:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                    mask_ids.add(extract_id(file))

    # Find matching IDs
    matched_ids = list(image_ids.intersection(mask_ids))

    print(f"Total unique image IDs: {len(image_ids)}")
    print(f"Total unique mask IDs: {len(mask_ids)}")
    print(f"Matched pairs: {len(matched_ids)}")

    if len(matched_ids) > 0:
        print(f"Sample matched IDs: {sorted(matched_ids)[:5]}")
    else:
        print("❌ No matching pairs found!")
        print("Sample image IDs:", sorted(list(image_ids))[:5])
        print("Sample mask IDs:", sorted(list(mask_ids))[:5])

    return matched_ids, image_dirs[0] if image_dirs else None, mask_dirs[0] if mask_dirs else None

# Step 4: Basic data quality check
def check_data_quality(matched_ids, image_dir, mask_dir):
    """Check basic data quality metrics"""
    print("\n4. DATA QUALITY ASSESSMENT")
    print("-" * 40)

    if not matched_ids:
        print("❌ No matched pairs to analyze")
        return

    print(f"Analyzing {min(10, len(matched_ids))} sample pairs...")

    image_info = []
    mask_info = []

    for i, id_ in enumerate(matched_ids[:10]):
        try:
            # Find image file
            image_file = None
            for ext in ['.png', '.jpg', '.jpeg', '.tif', '.tiff']:
                potential_file = f"{id_}{ext}"
                if os.path.exists(os.path.join(image_dir, potential_file)):
                    image_file = potential_file
                    break

            # Find mask file
            mask_file = None
            for ext in ['.png', '.jpg', '.jpeg', '.tif', '.tiff']:
                potential_file = f"{id_}_classimg_nonconvex{ext}"
                if os.path.exists(os.path.join(mask_dir, potential_file)):
                    mask_file = potential_file
                    break

            if image_file and mask_file:
                # Load and analyze image
                img = Image.open(os.path.join(image_dir, image_file))
                mask = Image.open(os.path.join(mask_dir, mask_file))

                image_info.append({
                    'id': id_,
                    'size': img.size,
                    'mode': img.mode,
                    'format': img.format
                })

                mask_info.append({
                    'id': id_,
                    'size': mask.size,
                    'mode': mask.mode,
                    'unique_values': len(set(np.array(mask).flatten()))
                })

        except Exception as e:
            print(f"Error processing {id_}: {e}")

    # Summarize findings
    if image_info:
        print("✅ Image Analysis:")
        sizes = [info['size'] for info in image_info]
        modes = [info['mode'] for info in image_info]
        print(f"  - Image sizes: {Counter(sizes).most_common()}")
        print(f"  - Color modes: {Counter(modes).most_common()}")

        print("\n✅ Mask Analysis:")
        mask_sizes = [info['size'] for info in mask_info]
        unique_vals = [info['unique_values'] for info in mask_info]
        print(f"  - Mask sizes: {Counter(mask_sizes).most_common()}")
        print(f"  - Unique values in masks: {Counter(unique_vals).most_common()}")

        # Check if image and mask sizes match
        size_matches = sum(1 for i, m in zip(image_info, mask_info) if i['size'] == m['size'])
        print(f"  - Size matches: {size_matches}/{len(image_info)} pairs")

# Main execution
def main():
    try:
        # Step 1: Explore directory structure
        explore_directory_structure()

        # Step 2: Identify dataset files
        image_dirs, mask_dirs, all_files = identify_dataset_files()

        # Step 3: Match images and masks
        matched_ids, image_dir, mask_dir = match_images_and_masks(image_dirs, mask_dirs, all_files)

        # Step 4: Check data quality
        check_data_quality(matched_ids, image_dir, mask_dir)

        # Return the matched_ids for further use
        if matched_ids:
            print(f"\n🎉 SUCCESS: Found {len(matched_ids)} matched image-mask pairs!")
            print("You can now use the 'matched_ids' variable in your analysis.")
            return matched_ids, image_dir, mask_dir
        else:
            print("\n❌ No matched pairs found. Please check your data structure.")
            return [], None, None

    except Exception as e:
        print(f"❌ Error in main execution: {e}")
        return [], None, None

# Run the analysis
if __name__ == "__main__":
    matched_ids, image_dir, mask_dir = main()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

print("=== Gleason 2019 Dataset Initial Exploration ===\n")

# Step 1: Check directory structure and available files
def explore_directory_structure():
    """Explore the dataset directory structure"""
    print("1. DIRECTORY STRUCTURE ANALYSIS")
    print("-" * 40)

    # List all directories and files in current path
    current_files = os.listdir('.')
    print(f"Files/folders in current directory: {len(current_files)}")
    for item in sorted(current_files):
        if os.path.isdir(item):
            print(f"📁 {item}/")
            # List contents of each directory
            try:
                sub_items = os.listdir(item)
                print(f"   Contains {len(sub_items)} items")
                if len(sub_items) <= 10:  # Show contents if not too many
                    for sub_item in sorted(sub_items)[:5]:
                        print(f"   - {sub_item}")
                    if len(sub_items) > 5:
                        print(f"   ... and {len(sub_items)-5} more items")
            except PermissionError:
                print("   (Permission denied)")
        else:
            print(f"📄 {item}")
    print()

# Step 2: Identify image and mask files
def identify_dataset_files():
    """Identify and match image files with their corresponding masks"""
    print("2. DATASET FILES IDENTIFICATION")
    print("-" * 40)

    # Check all directories for potential images and masks
    image_dirs = []
    mask_dirs = []

    # Check each directory
    for dir_name in os.listdir('.'):
        if os.path.isdir(dir_name) and not dir_name.startswith('.'):
            try:
                files = os.listdir(dir_name)
                if files:  # Only check non-empty directories
                    # Check if this directory contains images or masks based on filenames
                    sample_files = files[:10]  # Check first 10 files

                    has_images = any(
                        f.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')) and
                        'classimg' not in f.lower() and 'mask' not in f.lower()
                        for f in sample_files
                    )

                    has_masks = any(
                        'classimg' in f.lower() or 'mask' in f.lower()
                        for f in sample_files
                    )

                    if has_images:
                        image_dirs.append(dir_name)
                    if has_masks:
                        mask_dirs.append(dir_name)

            except Exception as e:
                print(f"Error reading {dir_name}: {e}")

    print(f"Potential image directories: {image_dirs}")
    print(f"Potential mask directories: {mask_dirs}")

    # Get file lists from each directory
    all_files = {}
    for directory in image_dirs + mask_dirs:
        try:
            files = os.listdir(directory)
            all_files[directory] = files
            print(f"\n{directory}/ contains {len(files)} files")

            # Show file extensions
            extensions = {}
            for file in files:
                ext = os.path.splitext(file)[1].lower()
                extensions[ext] = extensions.get(ext, 0) + 1
            print(f"File extensions: {extensions}")

            # Show sample filenames
            print("Sample files:")
            for file in sorted(files)[:5]:
                print(f"  - {file}")

        except Exception as e:
            print(f"Error reading {directory}: {e}")

    return image_dirs, mask_dirs, all_files

# Step 3: Match images with their corresponding masks
def match_images_and_masks(image_dirs, mask_dirs, all_files):
    """Match image files with their corresponding mask files"""
    print("\n3. MATCHING IMAGES AND MASKS")
    print("-" * 40)

    # If no clear image directories found, check if we need to handle this differently
    if not image_dirs and mask_dirs:
        print("⚠️  Found masks but no obvious image directory.")
        print("Let's check if images might be mixed with masks or in a different format...")

        # Check if the mask directory also contains images
        for mask_dir in mask_dirs:
            if mask_dir in all_files:
                mask_files = all_files[mask_dir]
                # Look for files that might be images (not containing 'classimg')
                potential_images = [f for f in mask_files if 'classimg' not in f.lower()]
                if potential_images:
                    print(f"Found {len(potential_images)} potential image files in {mask_dir}/")
                    print("Sample potential images:", potential_images[:3])

    if not image_dirs or not mask_dirs:
        print("❌ Could not find both image and mask directories")
        print("Let's try a different approach - analyzing the file naming patterns...")

        # Alternative approach: analyze all files across directories
        all_image_files = []
        all_mask_files = []

        for directory in ['Gleason', 'Maps4_T']:  # Known directories from your output
            if directory in all_files:
                for file in all_files[directory]:
                    if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                        if 'classimg' in file.lower() or 'mask' in file.lower():
                            all_mask_files.append((directory, file))
                        else:
                            all_image_files.append((directory, file))

        print(f"Found {len(all_image_files)} potential image files")
        print(f"Found {len(all_mask_files)} potential mask files")

        if all_image_files:
            print("Sample image files:", [f[1] for f in all_image_files[:3]])
        if all_mask_files:
            print("Sample mask files:", [f[1] for f in all_mask_files[:3]])

        # Try to match based on naming patterns
        matched_pairs = []

        # Extract base IDs from all files
        def extract_id(filename):
            base = os.path.splitext(filename)[0]
            # Remove common suffixes
            for suffix in ['_classimg_nonconvex', '_mask', '_label', '_img']:
                if base.endswith(suffix):
                    base = base[:-len(suffix)]
            return base

        # Create dictionaries for easier matching
        image_dict = {}
        mask_dict = {}

        for directory, file in all_image_files:
            base_id = extract_id(file)
            image_dict[base_id] = (directory, file)

        for directory, file in all_mask_files:
            base_id = extract_id(file)
            mask_dict[base_id] = (directory, file)

        # Find matches
        matched_ids = []
        for base_id in image_dict:
            if base_id in mask_dict:
                matched_ids.append(base_id)
                matched_pairs.append({
                    'id': base_id,
                    'image_path': os.path.join(image_dict[base_id][0], image_dict[base_id][1]),
                    'mask_path': os.path.join(mask_dict[base_id][0], mask_dict[base_id][1])
                })

        print(f"Successfully matched {len(matched_pairs)} image-mask pairs!")
        if matched_pairs:
            print("Sample matches:")
            for pair in matched_pairs[:3]:
                print(f"  {pair['id']}: {pair['image_path']} <-> {pair['mask_path']}")

        return matched_ids, matched_pairs

    # Original matching logic for when directories are clearly separated
    def extract_id(filename):
        base = os.path.splitext(filename)[0]
        for suffix in ['_classimg_nonconvex', '_mask', '_label', '_img']:
            if base.endswith(suffix):
                base = base[:-len(suffix)]
        return base

    image_ids = set()
    mask_ids = set()

    for img_dir in image_dirs:
        if img_dir in all_files:
            for file in all_files[img_dir]:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                    image_ids.add(extract_id(file))

    for mask_dir in mask_dirs:
        if mask_dir in all_files:
            for file in all_files[mask_dir]:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                    mask_ids.add(extract_id(file))

    matched_ids = list(image_ids.intersection(mask_ids))

    print(f"Total unique image IDs: {len(image_ids)}")
    print(f"Total unique mask IDs: {len(mask_ids)}")
    print(f"Matched pairs: {len(matched_ids)}")

    if len(matched_ids) > 0:
        print(f"Sample matched IDs: {sorted(matched_ids)[:5]}")

        # Create matched pairs info
        matched_pairs = []
        for id_ in matched_ids:
            matched_pairs.append({
                'id': id_,
                'image_dir': image_dirs[0],
                'mask_dir': mask_dirs[0]
            })

        return matched_ids, matched_pairs
    else:
        print("❌ No matching pairs found!")
        print("Sample image IDs:", sorted(list(image_ids))[:5])
        print("Sample mask IDs:", sorted(list(mask_ids))[:5])
        return [], []

# Step 4: Basic data quality check
def check_data_quality(matched_pairs):
    """Check basic data quality metrics"""
    print("\n4. DATA QUALITY ASSESSMENT")
    print("-" * 40)

    if not matched_pairs:
        print("❌ No matched pairs to analyze")
        return

    print(f"Analyzing {min(10, len(matched_pairs))} sample pairs...")

    image_info = []
    mask_info = []

    for i, pair in enumerate(matched_pairs[:10]):
        try:
            # Load images based on the pair structure
            if 'image_path' in pair:
                # Direct path approach
                img = Image.open(pair['image_path'])
                mask = Image.open(pair['mask_path'])
            else:
                # Directory + ID approach
                id_ = pair['id']
                image_dir = pair['image_dir']
                mask_dir = pair['mask_dir']

                # Find image file
                image_file = None
                for ext in ['.png', '.jpg', '.jpeg', '.tif', '.tiff']:
                    potential_file = f"{id_}{ext}"
                    if os.path.exists(os.path.join(image_dir, potential_file)):
                        image_file = potential_file
                        break

                # Find mask file
                mask_file = None
                for ext in ['.png', '.jpg', '.jpeg', '.tif', '.tiff']:
                    potential_file = f"{id_}_classimg_nonconvex{ext}"
                    if os.path.exists(os.path.join(mask_dir, potential_file)):
                        mask_file = potential_file
                        break

                if not image_file or not mask_file:
                    continue

                img = Image.open(os.path.join(image_dir, image_file))
                mask = Image.open(os.path.join(mask_dir, mask_file))

            # Convert mask to numpy for analysis
            mask_array = np.array(mask)

            image_info.append({
                'id': pair['id'],
                'size': img.size,
                'mode': img.mode,
                'format': img.format
            })

            mask_info.append({
                'id': pair['id'],
                'size': mask.size,
                'mode': mask.mode,
                'unique_values': len(np.unique(mask_array)),
                'value_range': f"{mask_array.min()}-{mask_array.max()}"
            })

        except Exception as e:
            print(f"Error processing pair {i}: {e}")

    # Summarize findings
    if image_info:
        print("✅ Image Analysis:")
        sizes = [info['size'] for info in image_info]
        modes = [info['mode'] for info in image_info]
        print(f"  - Image sizes: {Counter(sizes).most_common()}")
        print(f"  - Color modes: {Counter(modes).most_common()}")

        print("\n✅ Mask Analysis:")
        mask_sizes = [info['size'] for info in mask_info]
        unique_vals = [info['unique_values'] for info in mask_info]
        value_ranges = [info['value_range'] for info in mask_info]
        print(f"  - Mask sizes: {Counter(mask_sizes).most_common()}")
        print(f"  - Unique values in masks: {Counter(unique_vals).most_common()}")
        print(f"  - Value ranges: {Counter(value_ranges).most_common()}")

        # Check if image and mask sizes match
        size_matches = sum(1 for i, m in zip(image_info, mask_info) if i['size'] == m['size'])
        print(f"  - Size matches: {size_matches}/{len(image_info)} pairs")

        return image_info, mask_info

    return [], []

# Main execution
def main():
    try:
        # Step 1: Explore directory structure
        explore_directory_structure()

        # Step 2: Identify dataset files
        image_dirs, mask_dirs, all_files = identify_dataset_files()

        # Step 3: Match images and masks
        matched_ids, matched_pairs = match_images_and_masks(image_dirs, mask_dirs, all_files)

        # Step 4: Check data quality
        image_info, mask_info = check_data_quality(matched_pairs)

        # Return the results for further use
        if matched_pairs:
            print(f"\n🎉 SUCCESS: Found {len(matched_pairs)} matched image-mask pairs!")
            print("You can now use the 'matched_ids' and 'matched_pairs' variables in your analysis.")

            # Create a simple matched_ids list for backward compatibility
            simple_matched_ids = [pair['id'] for pair in matched_pairs]

            return simple_matched_ids, matched_pairs, image_info, mask_info
        else:
            print("\n❌ No matched pairs found. Please check your data structure.")
            return [], [], [], []

    except Exception as e:
        print(f"❌ Error in main execution: {e}")
        import traceback
        traceback.print_exc()
        return [], [], [], []

# Run the analysis
if __name__ == "__main__":
    matched_ids, matched_pairs, image_info, mask_info = main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import pandas as pd
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Set up plotting
plt.style.use('default')
sns.set_palette("husl")
fig_size = (15, 10)

print("=== Comprehensive Gleason Dataset EDA ===\n")

# 1. CLASS DISTRIBUTION ANALYSIS
def analyze_class_distribution(matched_ids):
    """Analyze the distribution of Gleason grades across the dataset"""
    print("1. CLASS DISTRIBUTION ANALYSIS")
    print("-" * 50)

    grade_counts = {0: 0, 1: 0, 2: 0, 3: 0, 4: 0}
    pixel_counts = {0: 0, 1: 0, 2: 0, 3: 0, 4: 0}

    # Sample analysis on subset for speed
    sample_size = min(50, len(matched_ids))
    sample_ids = matched_ids[:sample_size]

    print(f"Analyzing class distribution on {sample_size} samples...")

    for i, id_ in enumerate(sample_ids):
        try:
            mask_path = f"Maps4_T/{id_}_classimg_nonconvex.png"
            mask = np.array(Image.open(mask_path))

            # Count unique values in this mask
            unique, counts = np.unique(mask, return_counts=True)

            for grade, count in zip(unique, counts):
                if grade in grade_counts:
                    pixel_counts[grade] += count
                    if count > 0:  # If this grade appears in this image
                        grade_counts[grade] += 1

            if (i + 1) % 10 == 0:
                print(f"Processed {i + 1}/{sample_size} samples...")

        except Exception as e:
            print(f"Error processing {id_}: {e}")

    # Create visualizations
    fig, axes = plt.subplots(2, 2, figsize=fig_size)

    # Image-level distribution
    grades = list(grade_counts.keys())
    image_counts = list(grade_counts.values())

    axes[0,0].bar(grades, image_counts, alpha=0.7, color='skyblue')
    axes[0,0].set_title('Number of Images Containing Each Grade')
    axes[0,0].set_xlabel('Gleason Grade')
    axes[0,0].set_ylabel('Number of Images')

    # Pixel-level distribution
    pixel_counts_list = list(pixel_counts.values())
    axes[0,1].bar(grades, pixel_counts_list, alpha=0.7, color='lightcoral')
    axes[0,1].set_title('Total Pixel Count by Grade')
    axes[0,1].set_xlabel('Gleason Grade')
    axes[0,1].set_ylabel('Total Pixels')
    axes[0,1].ticklabel_format(style='scientific', axis='y', scilimits=(0,0))

    # Percentage distribution
    total_pixels = sum(pixel_counts_list)
    percentages = [count/total_pixels*100 for count in pixel_counts_list]

    axes[1,0].pie(percentages, labels=[f'Grade {g}' for g in grades],
                  autopct='%1.1f%%', startangle=90)
    axes[1,0].set_title('Pixel Distribution by Grade (%)')

    # Class imbalance visualization
    axes[1,1].bar(grades, percentages, alpha=0.7, color='lightgreen')
    axes[1,1].set_title('Class Imbalance Analysis')
    axes[1,1].set_xlabel('Gleason Grade')
    axes[1,1].set_ylabel('Percentage of Total Pixels')
    axes[1,1].axhline(y=20, color='red', linestyle='--', alpha=0.7, label='Balanced (20%)')
    axes[1,1].legend()

    plt.tight_layout()
    plt.show()

    # Print summary statistics
    print("\n📊 Class Distribution Summary:")
    print(f"{'Grade':<8} {'Images':<8} {'Pixels':<12} {'Percentage':<12}")
    print("-" * 45)
    for grade in grades:
        print(f"{grade:<8} {grade_counts[grade]:<8} {pixel_counts[grade]:<12,} {percentages[grade]:<12.2f}%")

    # Identify class imbalance issues
    max_percentage = max(percentages)
    min_percentage = min([p for p in percentages if p > 0])
    imbalance_ratio = max_percentage / min_percentage if min_percentage > 0 else float('inf')

    print(f"\n⚠️  Class Imbalance Analysis:")
    print(f"   - Most common grade: {percentages.index(max_percentage)} ({max_percentage:.2f}%)")
    print(f"   - Least common grade: {percentages.index(min_percentage)} ({min_percentage:.2f}%)")
    print(f"   - Imbalance ratio: {imbalance_ratio:.1f}:1")

    if imbalance_ratio > 10:
        print("   - ⚠️ Severe class imbalance detected! Consider weighted loss or resampling.")
    elif imbalance_ratio > 5:
        print("   - ⚠️ Moderate class imbalance detected. Monitor performance carefully.")
    else:
        print("   - ✅ Class distribution is relatively balanced.")

    return grade_counts, pixel_counts, percentages

# 2. IMAGE STATISTICS AND STAINING ANALYSIS
def analyze_image_statistics(matched_ids):
    """Analyze image statistics and staining variability"""
    print("\n2. IMAGE STATISTICS AND STAINING ANALYSIS")
    print("-" * 50)

    sample_size = min(30, len(matched_ids))
    sample_ids = matched_ids[:sample_size]

    image_stats = []

    print(f"Analyzing image statistics on {sample_size} samples...")

    for i, id_ in enumerate(sample_ids):
        try:
            image_path = f"Gleason/{id_} (1).jpg"
            if not os.path.exists(image_path):
                # Try alternative naming
                image_path = f"Gleason/{id_}.jpg"

            img = np.array(Image.open(image_path))

            # Calculate statistics for each channel
            stats = {
                'id': id_,
                'height': img.shape[0],
                'width': img.shape[1],
                'mean_r': np.mean(img[:,:,0]),
                'mean_g': np.mean(img[:,:,1]),
                'mean_b': np.mean(img[:,:,2]),
                'std_r': np.std(img[:,:,0]),
                'std_g': np.std(img[:,:,1]),
                'std_b': np.std(img[:,:,2]),
                'brightness': np.mean(img),
                'contrast': np.std(img)
            }
            image_stats.append(stats)

        except Exception as e:
            print(f"Error processing {id_}: {e}")

    if not image_stats:
        print("❌ No images could be processed")
        return

    df_stats = pd.DataFrame(image_stats)

    # Create visualizations
    fig, axes = plt.subplots(2, 3, figsize=fig_size)

    # RGB channel means
    axes[0,0].hist([df_stats['mean_r'], df_stats['mean_g'], df_stats['mean_b']],
                   bins=15, alpha=0.7, label=['Red', 'Green', 'Blue'])
    axes[0,0].set_title('RGB Channel Mean Values')
    axes[0,0].set_xlabel('Mean Pixel Value')
    axes[0,0].set_ylabel('Frequency')
    axes[0,0].legend()

    # RGB channel standard deviations
    axes[0,1].hist([df_stats['std_r'], df_stats['std_g'], df_stats['std_b']],
                   bins=15, alpha=0.7, label=['Red', 'Green', 'Blue'])
    axes[0,1].set_title('RGB Channel Standard Deviations')
    axes[0,1].set_xlabel('Standard Deviation')
    axes[0,1].set_ylabel('Frequency')
    axes[0,1].legend()

    # Brightness distribution
    axes[0,2].hist(df_stats['brightness'], bins=15, alpha=0.7, color='orange')
    axes[0,2].set_title('Overall Brightness Distribution')
    axes[0,2].set_xlabel('Average Brightness')
    axes[0,2].set_ylabel('Frequency')

    # Contrast distribution
    axes[1,0].hist(df_stats['contrast'], bins=15, alpha=0.7, color='purple')
    axes[1,0].set_title('Contrast Distribution')
    axes[1,0].set_xlabel('Contrast (Std Dev)')
    axes[1,0].set_ylabel('Frequency')

    # Brightness vs Contrast scatter
    axes[1,1].scatter(df_stats['brightness'], df_stats['contrast'], alpha=0.6)
    axes[1,1].set_title('Brightness vs Contrast')
    axes[1,1].set_xlabel('Brightness')
    axes[1,1].set_ylabel('Contrast')

    # RGB correlation
    rgb_corr = df_stats[['mean_r', 'mean_g', 'mean_b']].corr()
    sns.heatmap(rgb_corr, annot=True, cmap='coolwarm', center=0, ax=axes[1,2])
    axes[1,2].set_title('RGB Channel Correlation')

    plt.tight_layout()
    plt.show()

    # Print summary statistics
    print("📊 Image Statistics Summary:")
    numeric_cols = ['brightness', 'contrast', 'mean_r', 'mean_g', 'mean_b', 'std_r', 'std_g', 'std_b']
    summary = df_stats[numeric_cols].describe()
    print(summary.round(2))

    # Staining variability analysis
    cv_brightness = df_stats['brightness'].std() / df_stats['brightness'].mean()
    cv_red = df_stats['mean_r'].std() / df_stats['mean_r'].mean()
    cv_green = df_stats['mean_g'].std() / df_stats['mean_g'].mean()
    cv_blue = df_stats['mean_b'].std() / df_stats['mean_b'].mean()

    print(f"\n🎨 Staining Variability Analysis:")
    print(f"   - Brightness CV: {cv_brightness:.3f}")
    print(f"   - Red channel CV: {cv_red:.3f}")
    print(f"   - Green channel CV: {cv_green:.3f}")
    print(f"   - Blue channel CV: {cv_blue:.3f}")

    if max(cv_red, cv_green, cv_blue) > 0.15:
        print("   - ⚠️ High staining variability detected! Consider stain normalization.")
    else:
        print("   - ✅ Staining variability is within acceptable range.")

    return df_stats

# 3. SPATIAL PATTERN ANALYSIS
def analyze_spatial_patterns(matched_ids):
    """Analyze spatial patterns in Gleason grades"""
    print("\n3. SPATIAL PATTERN ANALYSIS")
    print("-" * 50)

    sample_size = min(20, len(matched_ids))
    sample_ids = matched_ids[:sample_size]

    print(f"Analyzing spatial patterns on {sample_size} samples...")

    # Analyze grade co-occurrence patterns
    grade_transitions = np.zeros((5, 5))  # 5x5 matrix for grades 0-4

    for i, id_ in enumerate(sample_ids):
        try:
            mask_path = f"Maps4_T/{id_}_classimg_nonconvex.png"
            mask = np.array(Image.open(mask_path))

            # Analyze horizontal transitions
            for row in mask:
                for j in range(len(row)-1):
                    current_grade = int(row[j])
                    next_grade = int(row[j+1])
                    if 0 <= current_grade <= 4 and 0 <= next_grade <= 4:
                        grade_transitions[current_grade][next_grade] += 1

            # Analyze vertical transitions
            for col in range(mask.shape[1]):
                for j in range(mask.shape[0]-1):
                    current_grade = int(mask[j, col])
                    next_grade = int(mask[j+1, col])
                    if 0 <= current_grade <= 4 and 0 <= next_grade <= 4:
                        grade_transitions[current_grade][next_grade] += 1

        except Exception as e:
            print(f"Error processing {id_}: {e}")

    # Normalize transition matrix
    transition_probs = grade_transitions / (grade_transitions.sum(axis=1, keepdims=True) + 1e-8)

    # Visualize spatial patterns
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Transition matrix heatmap
    sns.heatmap(transition_probs, annot=True, fmt='.3f', cmap='Blues',
                xticklabels=[f'Grade {i}' for i in range(5)],
                yticklabels=[f'Grade {i}' for i in range(5)],
                ax=axes[0])
    axes[0].set_title('Grade Transition Probabilities')
    axes[0].set_xlabel('Next Grade')
    axes[0].set_ylabel('Current Grade')

    # Diagonal dominance analysis
    diagonal_strength = np.diag(transition_probs)
    axes[1].bar(range(5), diagonal_strength, alpha=0.7, color='green')
    axes[1].set_title('Spatial Coherence by Grade')
    axes[1].set_xlabel('Gleason Grade')
    axes[1].set_ylabel('Self-Transition Probability')
    axes[1].set_xticks(range(5))
    axes[1].axhline(y=0.5, color='red', linestyle='--', alpha=0.7, label='Random (0.5)')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    print("🗺️  Spatial Pattern Analysis:")
    print(f"   - Average spatial coherence: {np.mean(diagonal_strength):.3f}")

    if np.mean(diagonal_strength) > 0.7:
        print("   - ✅ High spatial coherence - grades tend to cluster together")
    elif np.mean(diagonal_strength) > 0.5:
        print("   - ⚠️ Moderate spatial coherence")
    else:
        print("   - ❌ Low spatial coherence - grades are spatially scattered")

    return transition_probs

# 4. SAMPLE VISUALIZATION
def visualize_samples(matched_ids, num_samples=6):
    """Visualize sample images with their corresponding masks"""
    print(f"\n4. SAMPLE VISUALIZATION ({num_samples} examples)")
    print("-" * 50)

    sample_ids = matched_ids[:num_samples]

    fig, axes = plt.subplots(3, num_samples, figsize=(3*num_samples, 9))
    if num_samples == 1:
        axes = axes.reshape(-1, 1)

    for i, id_ in enumerate(sample_ids):
        try:
            # Load image
            image_path = f"Gleason/{id_} (1).jpg"
            if not os.path.exists(image_path):
                image_path = f"Gleason/{id_}.jpg"

            img = Image.open(image_path)

            # Load mask
            mask_path = f"Maps4_T/{id_}_classimg_nonconvex.png"
            mask = Image.open(mask_path)
            mask_array = np.array(mask)

            # Original image
            axes[0,i].imshow(img)
            axes[0,i].set_title(f'Original\n{id_}', fontsize=10)
            axes[0,i].axis('off')

            # Mask
            axes[1,i].imshow(mask_array, cmap='viridis', vmin=0, vmax=4)
            axes[1,i].set_title('Mask', fontsize=10)
            axes[1,i].axis('off')

            # Overlay
            overlay = np.array(img)
            mask_colored = plt.cm.viridis(mask_array / 4.0)[:,:,:3]
            overlay_blend = 0.7 * overlay/255.0 + 0.3 * mask_colored
            axes[2,i].imshow(overlay_blend)
            axes[2,i].set_title('Overlay', fontsize=10)
            axes[2,i].axis('off')

            # Add grade information
            unique_grades = np.unique(mask_array)
            grade_text = f"Grades: {sorted(unique_grades)}"
            axes[1,i].text(0.02, 0.98, grade_text, transform=axes[1,i].transAxes,
                          fontsize=8, verticalalignment='top',
                          bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

        except Exception as e:
            print(f"Error visualizing {id_}: {e}")
            for row in range(3):
                axes[row,i].text(0.5, 0.5, 'Error\nloading',
                               ha='center', va='center', transform=axes[row,i].transAxes)
                axes[row,i].axis('off')

    plt.tight_layout()
    plt.show()

# RUN ALL ANALYSES
def run_comprehensive_eda(matched_ids):
    """Run all EDA analyses"""
    print(f"Starting comprehensive EDA on {len(matched_ids)} matched pairs...\n")

    # Run all analysis functions
    try:
        grade_counts, pixel_counts, percentages = analyze_class_distribution(matched_ids)
        df_stats = analyze_image_statistics(matched_ids)
        transition_probs = analyze_spatial_patterns(matched_ids)
        visualize_samples(matched_ids, num_samples=6)

        print("\n" + "="*60)
        print("🎉 COMPREHENSIVE EDA COMPLETED SUCCESSFULLY!")
        print("="*60)

        # Final recommendations
        print("\n📋 RECOMMENDATIONS FOR MODEL DEVELOPMENT:")

        # Class imbalance recommendations
        max_percentage = max(percentages)
        min_percentage = min([p for p in percentages if p > 0])
        imbalance_ratio = max_percentage / min_percentage if min_percentage > 0 else float('inf')

        if imbalance_ratio > 10:
            print("1. Class Imbalance: Use weighted loss function or focal loss")
            print("   - Consider oversampling minority classes")
            print("   - Use stratified sampling for train/val splits")

        # Data augmentation recommendations
        print("2. Data Augmentation: Recommended techniques:")
        print("   - Rotation (90°, 180°, 270°)")
        print("   - Horizontal/vertical flipping")
        print("   - Elastic deformation (moderate)")
        print("   - Color jittering for stain robustness")

        # Architecture recommendations
        print("3. Model Architecture:")
        print("   - U-Net with skip connections for segmentation")
        print("   - Consider multi-scale inputs for large images")
        print("   - Use patch-based training for memory efficiency")

        # Evaluation recommendations
        print("4. Evaluation Metrics:")
        print("   - Dice coefficient for segmentation quality")
        print("   - Per-class IoU for individual grade performance")
        print("   - Confusion matrix for grade classification errors")

    except Exception as e:
        print(f"❌ Error during EDA: {e}")
        import traceback
        traceback.print_exc()

# Execute the comprehensive EDA
print("Ready to run comprehensive EDA!")
print("Execute: run_comprehensive_eda(matched_ids)")
```

```python
# 🚀 FAST-TRACK GLEASON IMPLEMENTATION PLAN
# Based on your excellent EDA results

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import numpy as np
from PIL import Image
import os
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

print("🎯 FAST-TRACK IMPLEMENTATION PLAN")
print("="*50)

# STEP 1: IMMEDIATE SETUP (5 minutes)
def setup_project_structure():
    """Create organized project structure"""
    directories = [
        'gleason_project',
        'gleason_project/data',
        'gleason_project/models',
        'gleason_project/utils',
        'gleason_project/configs',
        'gleason_project/results'
    ]

    for dir_path in directories:
        os.makedirs(dir_path, exist_ok=True)
        print(f"✅ Created: {dir_path}")

    print("\n📁 Project structure ready!")

# STEP 2: ROBUST DATA SPLITS (10 minutes)
def create_intelligent_splits(matched_ids):
    """Create robust splits that handle class imbalance"""
    print("🔄 Creating intelligent train/val/test splits...")

    # Analyze dominant grade in each image
    image_metadata = []

    sample_size = min(100, len(matched_ids))  # Use larger sample
    print(f"Analyzing {sample_size} samples for splitting...")

    for id_ in matched_ids[:sample_size]:
        try:
            mask = np.array(Image.open(f"Maps4_T/{id_}_classimg_nonconvex.png"))

            # Get grade distribution (excluding background)
            tissue_mask = mask[mask > 0]
            if len(tissue_mask) > 0:
                unique, counts = np.unique(tissue_mask, return_counts=True)
                dominant_grade = unique[np.argmax(counts)]

                # Calculate grade diversity
                num_grades = len(unique)

                image_metadata.append({
                    'id': id_,
                    'dominant_grade': dominant_grade,
                    'num_grades': num_grades,
                    'total_tissue': len(tissue_mask)
                })
        except Exception as e:
            continue

    # Check class distribution
    from collections import Counter
    label_counts = Counter([item['dominant_grade'] for item in image_metadata])
    print(f"📊 Class distribution: {dict(label_counts)}")

    # Filter out classes with too few samples for stratification
    min_samples_per_class = 2
    valid_classes = [grade for grade, count in label_counts.items() if count >= min_samples_per_class]

    # Filter metadata to only include valid classes
    filtered_metadata = [item for item in image_metadata if item['dominant_grade'] in valid_classes]

    if len(filtered_metadata) < 6:  # Need at least 6 samples for 70/15/15 split
        print("⚠️  Too few samples for stratified split. Using random split...")
        # Fall back to simple random split
        ids = [item['id'] for item in image_metadata]

        # Simple 70/20/10 split without stratification
        n_train = int(0.7 * len(ids))
        n_val = int(0.2 * len(ids))

        np.random.seed(42)
        shuffled_ids = np.random.permutation(ids)

        train_ids = shuffled_ids[:n_train].tolist()
        val_ids = shuffled_ids[n_train:n_train+n_val].tolist()
        test_ids = shuffled_ids[n_train+n_val:].tolist()

    else:
        # Use stratified split on filtered data
        ids = [item['id'] for item in filtered_metadata]
        labels = [item['dominant_grade'] for item in filtered_metadata]

        try:
            # 70/30 split first
            train_ids, temp_ids, train_labels, temp_labels = train_test_split(
                ids, labels, test_size=0.3, stratify=labels, random_state=42
            )

            # Split temp into val/test (50/50)
            val_ids, test_ids = train_test_split(
                temp_ids, test_size=0.5, stratify=temp_labels, random_state=42
            )

        except ValueError:
            print("⚠️  Stratification failed. Using random split...")
            # Fall back to random split
            np.random.seed(42)
            shuffled_ids = np.random.permutation(ids)

            n_train = int(0.7 * len(ids))
            n_val = int(0.15 * len(ids))

            train_ids = shuffled_ids[:n_train].tolist()
            val_ids = shuffled_ids[n_train:n_train+n_val].tolist()
            test_ids = shuffled_ids[n_train+n_val:].tolist()

    print(f"✅ Train: {len(train_ids)} samples")
    print(f"✅ Val: {len(val_ids)} samples")
    print(f"✅ Test: {len(test_ids)} samples")

    # Verify splits
    all_split_ids = set(train_ids + val_ids + test_ids)
    print(f"✅ Total unique IDs in splits: {len(all_split_ids)}")

    return train_ids, val_ids, test_ids, image_metadata

# STEP 3: OPTIMIZED DATA PIPELINE (15 minutes)
class GleasonDataset(Dataset):
    """Efficient dataset for patch-based training"""

    def __init__(self, image_ids, patch_size=512, patches_per_image=16,
                 mode='train', transform=None):
        self.image_ids = image_ids
        self.patch_size = patch_size
        self.patches_per_image = patches_per_image
        self.mode = mode
        self.transform = transform

        # Pre-compute valid patches to avoid runtime overhead
        self.patches = self._extract_patch_locations()

    def _extract_patch_locations(self):
        """Pre-compute patch locations for efficient training"""
        patches = []

        print(f"🔄 Pre-computing patch locations for {len(self.image_ids)} images...")

        for id_ in self.image_ids:
            try:
                mask_path = f"Maps4_T/{id_}_classimg_nonconvex.png"
                mask = np.array(Image.open(mask_path))

                h, w = mask.shape
                step = self.patch_size // 2  # 50% overlap

                patch_count = 0
                for y in range(0, h - self.patch_size, step):
                    for x in range(0, w - self.patch_size, step):
                        # Check if patch contains tissue (non-zero values)
                        patch_mask = mask[y:y+self.patch_size, x:x+self.patch_size]

                        if np.sum(patch_mask > 0) > (self.patch_size**2 * 0.1):  # >10% tissue
                            patches.append({
                                'id': id_,
                                'x': x, 'y': y,
                                'has_tissue': True
                            })
                            patch_count += 1

                            if patch_count >= self.patches_per_image:
                                break
                    if patch_count >= self.patches_per_image:
                        break

            except Exception as e:
                print(f"❌ Error processing {id_}: {e}")
                continue

        print(f"✅ Found {len(patches)} valid patches")
        return patches

    def __len__(self):
        return len(self.patches)

    def __getitem__(self, idx):
        patch_info = self.patches[idx]

        try:
            # Load image patch
            image_path = f"Gleason/{patch_info['id']} (1).jpg"
            if not os.path.exists(image_path):
                image_path = f"Gleason/{patch_info['id']}.jpg"

            image = Image.open(image_path)
            image = np.array(image)

            # Load mask patch
            mask_path = f"Maps4_T/{patch_info['id']}_classimg_nonconvex.png"
            mask = np.array(Image.open(mask_path))

            # Extract patches
            x, y = patch_info['x'], patch_info['y']
            image_patch = image[y:y+self.patch_size, x:x+self.patch_size]
            mask_patch = mask[y:y+self.patch_size, x:x+self.patch_size]

            # Apply transforms
            if self.transform:
                # Convert to PIL for transforms
                image_patch = Image.fromarray(image_patch)
                image_patch = self.transform(image_patch)
            else:
                image_patch = torch.from_numpy(image_patch).permute(2, 0, 1).float() / 255.0

            mask_patch = torch.from_numpy(mask_patch).long()

            return image_patch, mask_patch

        except Exception as e:
            # Return dummy data on error
            dummy_image = torch.zeros(3, self.patch_size, self.patch_size)
            dummy_mask = torch.zeros(self.patch_size, self.patch_size, dtype=torch.long)
            return dummy_image, dummy_mask

# STEP 4: SMART AUGMENTATIONS (5 minutes)
def get_train_transforms():
    """Optimized augmentations based on your EDA"""
    return transforms.Compose([
        transforms.RandomRotation(degrees=[0, 90, 180, 270]),  # Histology-appropriate
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05),  # Light jittering
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet stats
    ])

def get_val_transforms():
    """Validation transforms - no augmentation"""
    return transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

# STEP 5: CLASS WEIGHT CALCULATION (5 minutes)
def calculate_class_weights(matched_ids, sample_size=30):
    """Calculate class weights for handling imbalance"""
    print("⚖️  Calculating class weights...")

    grade_counts = {0: 0, 1: 0, 3: 0, 4: 0}  # Skip grade 2 for now

    for id_ in matched_ids[:sample_size]:
        try:
            mask = np.array(Image.open(f"Maps4_T/{id_}_classimg_nonconvex.png"))
            unique, counts = np.unique(mask, return_counts=True)

            for grade, count in zip(unique, counts):
                if grade in grade_counts:
                    grade_counts[grade] += count
        except:
            continue

    # Calculate inverse frequency weights
    total_pixels = sum(grade_counts.values())
    num_classes = len([c for c in grade_counts.values() if c > 0])

    class_weights = {}
    for grade, count in grade_counts.items():
        if count > 0:
            class_weights[grade] = total_pixels / (num_classes * count)
        else:
            class_weights[grade] = 1.0

    print("📊 Class weights:")
    for grade, weight in class_weights.items():
        print(f"   Grade {grade}: {weight:.3f}")

    return class_weights

# STEP 6: QUICK MODEL DEFINITION (10 minutes)
class SimpleUNet(nn.Module):
    """Lightweight U-Net for quick start"""

    def __init__(self, in_channels=3, num_classes=5):
        super().__init__()

        # Encoder
        self.enc1 = self._conv_block(in_channels, 64)
        self.enc2 = self._conv_block(64, 128)
        self.enc3 = self._conv_block(128, 256)
        self.enc4 = self._conv_block(256, 512)

        # Decoder
        self.dec4 = self._upconv_block(512, 256)
        self.dec3 = self._upconv_block(256, 128)
        self.dec2 = self._upconv_block(128, 64)
        self.dec1 = nn.Conv2d(64, num_classes, 1)

        self.pool = nn.MaxPool2d(2)

    def _conv_block(self, in_ch, out_ch):
        return nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )

    def _upconv_block(self, in_ch, out_ch):
        return nn.Sequential(
            nn.ConvTranspose2d(in_ch, out_ch, 2, stride=2),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))

        # Decoder with skip connections
        d4 = self.dec4(e4) + e3
        d3 = self.dec3(d4) + e2
        d2 = self.dec2(d3) + e1
        d1 = self.dec1(d2)

        return d1

# EXECUTION PLAN
def execute_fast_track_plan(matched_ids):
    """Execute the complete fast-track plan"""
    print("🚀 EXECUTING FAST-TRACK PLAN")
    print("="*40)

    # Step 1: Setup
    setup_project_structure()

    # Step 2: Data splits
    train_ids, val_ids, test_ids, metadata = create_intelligent_splits(matched_ids)

    # Step 3: Class weights
    class_weights = calculate_class_weights(matched_ids)

    # Step 4: Create datasets (small scale for testing)
    print("\n🔄 Creating datasets...")
    train_dataset = GleasonDataset(
        train_ids[:5],  # Start small for testing
        patch_size=512,
        patches_per_image=8,
        mode='train',
        transform=get_train_transforms()
    )

    print(f"✅ Training dataset: {len(train_dataset)} patches")

    # Step 5: Test data loading
    print("\n🔄 Testing data loading...")
    train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, num_workers=0)

    try:
        batch_images, batch_masks = next(iter(train_loader))
        print(f"✅ Data loading successful!")
        print(f"   Image batch shape: {batch_images.shape}")
        print(f"   Mask batch shape: {batch_masks.shape}")
        print(f"   Unique mask values: {torch.unique(batch_masks)}")

    except Exception as e:
        print(f"❌ Data loading error: {e}")

    # Step 6: Model setup
    print("\n🔄 Setting up model...")
    model = SimpleUNet(in_channels=3, num_classes=5)
    total_params = sum(p.numel() for p in model.parameters())
    print(f"✅ Model created: {total_params:,} parameters")

    print("\n🎉 FAST-TRACK SETUP COMPLETE!")
    print("="*40)
    print("📋 NEXT STEPS:")
    print("1. ✅ Run: execute_fast_track_plan(matched_ids)")
    print("2. 🔧 Start training loop implementation")
    print("3. 📊 Monitor training metrics")
    print("4. 🚀 Scale up to full dataset")

    return {
        'train_ids': train_ids,
        'val_ids': val_ids,
        'test_ids': test_ids,
        'class_weights': class_weights,
        'model': model,
        'train_dataset': train_dataset
    }

# READY TO EXECUTE!
print("🎯 READY TO START YOUR GLEASON PROJECT!")
print("Run: results = execute_fast_track_plan(matched_ids)")
```
